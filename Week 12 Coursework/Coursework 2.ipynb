<h2> Problem </h2>

In this coursework, we will build on the concept of grids that we discussed in Coursework 1.  In that coursework, we built a file called <samp>grids.py</samp> that included a definition for a class <samp>Grid</samp>.  This class modelled grids of $n \times n$ blocks, where each block can either be vacant or occupied.  Included in that class were methods that allowed us to add and delete occupancies, giving us control over which blocks were occupied.

We will extend this idea to model three specific types of grids:

<ul>
    <li><b>Uniform Grids</b>:  We say that a grid is <b>uniform</b> if each of its rows and columns has <em>exactly</em> two occupancies.</li><br>
    <li><b>No-Three-in-Line Grids</b>: We say that a grid satisfies the <b>no-three-in-line</b> property if there are no instances of three occupancies lying in a straight line.  Note that these lines do not necessarily have to be horizontal or vertical, a straight line of <em>any</em> gradient would count.</li><br>
    <li><b>UNTiL Grids</b>: We say that a grid is an <b>UNTiL grid</b> if it is both uniform <em>and</em> satisfies the no-three-in-line property.</li>
</ul>

We aim to subclass <samp>Grid</samp> to model each of these more specific types of grid structures.  Much of our effort will have to go into implementing tests for uniformity and the no-three-in-line property.  If a grid fails these tests, we will have to raise an appropriate error, and so new exceptions must be constructed for these contexts.  We will construct these exceptions in a separate file and import them.

We also wish to analyse the time complexity of instance construction across these classes.  We will do this by initialising instances of each class with the same occupancy data, and plotting the runtimes of each construction in a single figure.  Since these tests are independent of the class definitions, the code to run them should also be held in a separate file.

In short: we will be constructing a package to model the four types of grid structure: standard, uniform, no-three-in-line, and UNTiL; giving tailor-made exceptions when an instance does not meet the required properties, and giving a test script to allow a user to compare runtimes of the initialisation routines of each class.
