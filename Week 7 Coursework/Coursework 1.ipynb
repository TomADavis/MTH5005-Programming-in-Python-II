<h2> Problem </h2>

This coursework will require us to construct a Python file called <samp>grids.py</samp>.  In this file, we will define a class called <samp>Grid</samp> that will model an $n \times n$ grid of blocks (for some $n \in \mathbb{N}$), where each block can either be vacant or occupied.

For example, consider the following grid, where $n = 3$:

<pre>
□ ■ ■
■ ■ □
■ □ ■
</pre>

The <samp>■</samp> character will denote an occupied block, and the <samp>□</samp> character will denote a vacant one.  We can build a coordinate system over these grids in the following way:

<ul>
    <li> The top row has $x$ coordinate $0$, and this coordinate increases by one as we move vertically down. </li><br>
    <li> The first column has $y$ coordinate $0$, and this increases by one as we move horizontally right. </li>
</ul>

With this in mind, the above grid has occupancies at coordinates $(0, 1)$, $(0, 2)$, $(1, 0)$, $(1, 1)$, $(2, 0)$ and $(2, 2)$.

We will build our <samp>Grid</samp> class so that each instance will be constructed with two pieces of data: an integer <samp>n</samp>, giving the dimensions of the square grid, and a list <samp>occupancies</samp>, where each element of this list is a tuple of length two, corresponding to the coordinates of an occupied block.

We begin with the definition of the class, and its constructor method <samp>\_\_init\_\_</samp> .

```python
class Grid:
    def __init__(self, n, occupancies):
        self._n = n
        self._occupancies = occupancies
        self._rows = [[False]*self._n for _ in range(self._n)]
        for row, col in self._occupancies:
            self._rows[row][col] = True
```

This will form the seed of our <samp>Grid</samp> class.  Create a new Python file called <samp>grids.py</samp>, and copy-and-paste this definition into it.  Apart from the addition of some much-needed documentation, we will not change this definition of <samp>\_\_init\_\_</samp>.  We will only modify the definition of <samp>Grid</samp> by adding more methods to it.

Note that under this class definition, the grid shown above could be modelled with the following instance:

```python
x = Grid(n = 3, occupancies = [(0, 1), (0, 2), (1, 0), (1, 1), (2, 0), (2, 2)])
```

Of course, without any other methods, there is little we can do with this instance other than read its attributes.

```python
print(x._n)
print(x._occupancies)
```

It is clear that these attributes are fed in directly from the arguments given to the constructor function upon creation.  More subtle, is the <samp>_rows</samp> attribute, which is generated from the occupancy data.

```python
print(x._rows)
```
